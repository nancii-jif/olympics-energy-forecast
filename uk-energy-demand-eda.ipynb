```python
uk_elec_demand['datetime'] = pd.to_datetime(uk_elec_demand.ELEXM_utc)
uk_elec_demand['date'] = uk_elec_demand.datetime.dt.date
uk_elec_demand['time'] = uk_elec_demand.datetime.dt.time
uk_elec_demand = uk_elec_demand[uk_elec_demand.datetime.dt.year >= 2009]
```

```python
uk_elec_demand = uk_elec_demand[['datetime', 'total_demand', 'date', 'time']]
```

```python
uk_elec_demand.describe()
```

```python
uk_demand_12 = uk_elec_demand[uk_elec_demand.datetime.dt.year == 2012]
```

```python
uk_demand_12
```

```python
daily_t5 = uk_demand_12.groupby('date')[['total_demand', 'time']].apply(lambda x: x.nlargest(5, columns='total_demand')).reset_index()
```

```python
daily_t5
```

```python
daily_t5.time = pd.to_datetime(daily_t5.time, format='%H:%M:%S').dt.time
daily_t5.time = daily_t5.time.apply(lambda x: x.hour + x.minute / 60.0)
daily_t5
```

```python
daily_t5.dtypes
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(daily_t5.time, daily_t5.total_demand, color='b', alpha=0.7)
ax.set_xlabel('time')
ax.set_ylabel('electricity demand')
ax.set_title('top 5 half-hour windows of highest electricity demand in the UK')
plt.show()
```

```python
daily = uk_elec_demand.groupby('date')[['total_demand', 'time']].apply(lambda x: x.nlargest(1, columns='total_demand')).reset_index()
daily['top1_ma7'] = daily.total_demand.rolling(window=7, min_periods=1).mean()
daily_sum = uk_elec_demand.groupby('date')['total_demand'].sum().reset_index()
daily = pd.merge(daily, daily_sum, on='date')
daily = daily.rename(columns={'total_demand_x' : 't1_demand', 'time' : 'time_of_t1_demand', 'top1_ma7' : 't1_ma7', 'total_demand_y' : 'total_demand'})
daily['sum_ma7'] = daily.total_demand.rolling(window=7, min_periods=1).mean()
```

```python
daily.date = pd.to_datetime(daily.date)
daily['year'] = daily.date.dt.year
daily['day_of_year'] = daily.date.dt.dayofyear
daily
```

```python
daily.dtypes
```

```python
fig, ax = plt.subplots(figsize=(20, 5))
# ax.plot(daily.date, daily.sum_ma7)
ax.plot(daily.date, daily.t1_ma7)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 5))
ax.plot(daily.date, daily.sum_ma7)
# ax.plot(daily.date, daily.t1_ma7)
plt.show()
```

```python
from scipy.signal import detrend
```

```python
dt_total_demand = detrend(daily.sum_ma7, type='linear')
detrended_sum_ma7 = pd.Series(dt_total_demand, index=daily.index)
detrended_sum_ma7 = detrended_sum_ma7.diff(365)
daily['detrended_sum_ma7'] = detrended_sum_ma7
plt.figure(figsize=(21, 5))
plt.plot(daily.date, daily.detrended_sum_ma7)
```

```python
plt.figure(figsize=(21, 3))
plt.plot(detrended_sum_ma7.diff(365))
```

```python
plt.figure(figsize=(21, 5))
plt.plot(daily.date, daily.sum_ma7)
```

```python
fig, ax = plt.subplots(figsize=(21, 3))
for year in range(2010, 2015):
    year_data = daily[daily.date.dt.year == year]
    ax.plot(year_data.day_of_year, year_data.detrended_sum_ma7, label=year)

plt.legend()
plt.show()
```
